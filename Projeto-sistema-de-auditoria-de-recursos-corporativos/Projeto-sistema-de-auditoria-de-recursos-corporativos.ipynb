```python
import time
from functools import wraps

def auditor(funcao):
    """
    Decorator de auditoria: monitora parâmetros, tempo de execução
    e gera logs operacionais no terminal.
    """
    @wraps(funcao)
    def wrapper(*args, **kwargs):
        print("\n" + "="*50)
        print("🔍 INICIANDO AUDITORIA DE ORÇAMENTO")
        print(f"Setores Ignorados: {args}")
        print(f"Parâmetros de Conversão: {kwargs}")

        inicio = time.time()
        resultado = funcao(*args, **kwargs)
        fim = time.time()

        tempo_total = (fim - inicio) * 1000 # Convertido para ms

        print(f"⏱️ Tempo de Processamento: {tempo_total:.4f} ms")
        print("✅ AUDITORIA CONCLUÍDA")
        print("="*50 + "\n")

        return resultado
    return wrapper

# Estrutura de dados simulando a hierarquia da Multinacional
empresa_data = {
    "Matriz": {
        "Tecnologia": {
            "Cloud": 150000.00,
            "Seguranca": 85000.00,
            "IA_Labs": {
                "Equipe_Alpha": 50000.00,
                "Equipe_Beta": 45000.00
            }
        },
        "Recursos_Humanos": {
            "Recrutamento": 30000.00,
            "Treinamento": 20000.00
        },
        "Financeiro": {
            "Contabilidade": 40000.00,
            "Tesouraria": 60000.00
        }
    }
}

@auditor
def calcular_orcamento(dados, *setores_ignorados, **cambio):
    """
    Navega recursivamente pela árvore de orçamentos, aplicando filtros
    e conversão monetária final.
    """
    soma_total = 0

    for chave, valor in dados.items():
        # Regra de Auditoria: Ignorar o setor e toda a sua ramificação
        if chave in setores_ignorados:
            continue

        if isinstance(valor, dict):
            # Chamada recursiva para mergulhar na árvore
            soma_total += calcular_orcamento(valor, *setores_ignorados)
        else:
            # Caso base: valor numérico (folha)
            soma_total += valor

    # Aplicação do multiplicador via kwargs (taxa_cambio)
    taxa = cambio.get("taxa_cambio", 1.0)
    return soma_total * taxa

if __name__ == "__main__":
    # Exemplo de uso: Ignorando o RH e convertendo para BRL com taxa fictícia
    moeda = "BRL"
    taxa_conversao = 5.20

    total_auditado = calcular_orcamento(
        empresa_data,
        "Recursos_Humanos",
        moeda_destino=moeda,
        taxa_cambio=taxa_conversao
    )

    print(f"💰 Orçamento Total Consolidado ({moeda}): {total_auditado:,.2f}")
```
